**Formating Numbers**
```
pi = 3.14159

print(f"{pi:.2f}")   # 2 decimal places → 3.14

```
**Padding and alignment**
```
num = 42

print(f"{num:5}")     # width 5 → '   42'
print(f"{num:<5}")    # left aligned → '42   '
print(f"{num:^5}")    # center aligned → ' 42  '
print(f"{num:05}")    # zero padded → '00042'
```
**Working with strings**

#Data types
* Data types define the type of value a variable can store
* Python is dynamically typed (no need to declare type explicitly)

**primitive** (single type of data)
   1. int
   2. float
   3. string
   4. boolien
   5. complex

**Non Primitive** (hybrid type of data)
   1. list
   2. tuple
   3. set
   4. dict

**Types of Data Types**
1. Numeric Types
* int → integers (e.g., 10, -5)
* float → decimal numbers (e.g., 3.14)
* complex → complex numbers (e.g., 2 + 3j)
2. Sequence Types
* str → string (text data)
* list → ordered, mutable collection
* tuple → ordered, immutable collection
3. Set Types
* set → unordered, mutable, no duplicates
* frozenset → unordered, immutable set
4. Mapping Type
* dict → collection of key-value pairs
5. Boolean Type
* bool → True or False
6. Binary Types
* bytes → immutable binary data
* bytearray → mutable binary data
* memoryview → memory view object

* Allows duplicate characters

#Operators
* Operators are symbols used to perform operations on variables and values
* Operators work on operands (values/variables)
* Precedence (order of execution) matters
* Can be used with different data types

**Types**
1. Arithematic (+, -, *, /, //, %, **)
2. Comparison (==, !=, >, <, >=, <=)
3. Assignment (=, +=, -=, /=, *=)
4. Logical (AND, OR, NOT)
5. Identity (is, is not)
6. Membership Operators(in, not in)
7. Bitwise Operators(&, |, ^, ~, <<, >>)

````python
#Arithematic Operators
x1 = 90
x2 = 70
print(x1 * x2)
print(x1 + x2)
print(x1 / x2)
print(x1 % x2)
print(x1 // x2)
print(x1 - x2)
print(x1 ** x2)

#Comparison Operatoes
print(x1 == x2)
print(x1 != x2)
print(x1 < x2)
print(x1 > x2)
print(x1 <= x2)
print(x1 >= x2)

#Assignment Operator
ass = 109
print(ass)
ass += 1
print(ass)
ass -= 9
print(ass)
ass *= 10
print(ass)
ass /= 100
print(ass)

#Logical Operators
print(x1 == 90 and x2 == 70)

print(ass == 109 or x1 == 50)

print(not ass > 100)

#Identity Operator
print(id(x1),id(x2))
print(x1 is x2)
print(x1 is not x2)
print(id(ass))
print(ass is ass)
````

#List[ ]

* A list is an ordered, mutable collection of elements
* Defined using square brackets []
* Can store multiple data types (int, float, string, etc.)
* Indexing and slicing allowed
* Allows duplicate values
* Mutable → elements can be added, removed, or modified

**List Operations**
* concatenation → +
* repetition → *
* membership → in / not in

**Common Methods**
* append(x) → add element at end
* extend(iterable) → add multiple elements
* insert(i, x) → insert at index
* remove(x) → remove first occurrence
* pop([i]) → remove element by index (default last)
* clear() → remove all elements
* index(x) → first index of value
* count(x) → count occurrences
* sort() → sort list
* reverse() → reverse list
* copy() → shallow copy

**Built-in Functions**
* len(l) → number of elements
* max(l), min(l) → largest/smallest
* sum(l) → sum of elements

**Nested Lists**
```
l = [[1, 2], [3, 4]]
print(l[0][1])   # access nested element
```
**List Comprehension**
```
squares = [x**2 for x in range(5)]
```

````python
l1 = ['AG',22,'Pune',1000000]
print(id(l1))
print(len(l1))
print(l1)
print(type(l1))
l1[3] = 2000000
print(l1)
print(id(l1))

#Indexing

print(l1[0])

#Slicing

print(l1[0::])
print(l1[3::])
print(l1[0::2])
print(l1[1:3:])
print(l1[::-1])
print(l1[-4:-1:])
print(l1[-1:-5:-1])
print(l1[-1::-1])
````

#Tuple()

* A tuple is a collection of elements that is ordered and immutable
* Written using parentheses ()
* Can store different data types (int, string, etc.)
* Indexing allowed (positive & negative)
* Slicing allowed
* Allows duplicate values
* Faster than lists and uses less memory

**Tuple Operations**
1. concatenation → +
2. repetition → *
3. membership → in / not in

```
t = (1, 2, 3)        # packing
a, b, c = t          # unpacking
```
**Functions**
* len(t) → number of elements
* max(t), min(t) → largest/smallest

**Immutability Note**
* You cannot modify tuple elements
* But if it contains mutable items (like lists), those can change

**Methods**
* count(x) → count occurrences
* index(x) → first index of value




````python
t =(10,20,30,40,50,60)
print(t)
print(type(t))
print(len(t))

#indexing
print(t[0])
print(t[-1])

#slicing
print(t[0::])
print(t[::-1])
print(t[2:4:])
print(t[0::2])

#methods
print(t.index(30))
print(t.count(60))
````

#For/While Loop

**for loop**

```
for variable_name in range(start , stop , step):
    statement
```
* It goes through each element one by one
* Executes the loop body for every element
* Range function generates values till given argument.
* Used when the number of iterations is known or fixed
* Works with any iterable (list, tuple, string, etc.)
* Commonly used with range() for numbers

**while loop**
```
initialization(e.g i=1)
while(condition):
  statememt
  increment(e.g i += 1)
```
* It checks a condition first
* If the condition is True, the loop runs
* It keeps repeating until the condition becomes False
* Used when the number of iterations is not fixed
* Be careful to update the condition, or it may create an infinite loop


**Control statements:**
1.  break → to exit the loop early
2.   continue → to skip the current iteration

#List Comprehension
* List comprehension is a concise way to create lists in Python by combining a loop and optional conditions in a single line.
* Short & readable
* Faster than loops
* Cleaner code

**syntax**
```
[expression for item in iterable if condition]
```
* expression → what you want to store in the list
* item → each element from the iterable
* iterable → collection (list, tuple, string, etc.)
* condition (optional) → filter elements

````python
li = [i for i in range(1,11)]
print(type(li))
li
````

#set{}
* A set is an unordered collection of unique elements (no duplicates).
* No duplicates
* Unordered (no indexing)
* Mutable (can change)
* Elements must be immutable (int, string, tuple)

**Syntax**
```
my_set = {1, 2, 3}
empty_set = set()   # NOT {}
```
**Common Methods**
* add() → Add element
* pop() → Removes random element
* remove() → Remove element (error if not found)
* update(iterable) → adds multiple elements
* clear() → Empty set
* discard() → No error if not found

**Set Operations**
* Union (a | b) → Combines all unique elements from both sets
* Intersection (a & b) → Elements common to both sets
* Difference (a - b) → Elements in a but not in b
* Symmetric Difference (a ^ b) → Elements in either set but not in both
* membership (in / not in) → Check if element present in set or not

**Frozen set**
* Immutable version of set
```
fs = frozenset([1, 2, 3])
```
**Set Comprehension**
```
{x**2 for x in range(5)}
```


````python
s = {100,200,300,400,500}
print(s)
s.add(600)
print(s)
s.pop()
print(s)
s.remove(300)
print(s)
s.update({700,800,900,999})
print(s)
s.clear()
print(s)
````

#Dictionary{key:value}

* A dictionary in Python is a built-in data type that stores data in key–value pairs.
* Each key maps to a value
* Keys must be unique and immutable (e.g., strings, numbers, tuples)
* Values can be of any data type
* Mutable → can be changed after creation
* Unordered (before Python 3.7), insertion-ordered (3.7+)
* Indexed by keys, not positions
* No duplicate keys allowed

**Syntax**
```
my_dict = {
    "name": "Alice",
    "age": 21,
    "is_student": True
}
```

**Common Methods**
* keys() → keys
* values() → values
* get() → safe access
* items() → pairs
* update() → add/update
* clear() → empty dictionary

**Looping Through Dictionary**
```
for key, value in my_dict.items():
    print(key, value)
```

````python
data = {'name':'ag','age':22,'city':'pune','univercity':'RTMNU'}
print(data)
print(type(data))
data['name']='aradhya'
print(data)
del data['univercity']
data
````

````python
print(data.keys())
print(data.values())
print(data.get('age'))
print(data.items())
data.update({'experience':5,'salary':60000})
print(data)
data.clear()
print(data)
````

#User Defined Function

* A user-defined function is a block of reusable code created by the programmer using the def keyword. It performs a specific task and can be called whenever needed.
* def → keyword to define a function
* Function name → identifier of the function
* Parameters → inputs to the function
* Return statement → sends output back (optional)
* Docstring → describes the function (optional)

**Syntax**
```
def function_name(parameters):
    """optional docstring"""
    statements
    return value
```

# *Args & **Kwargs
***args (Non-keyword variable arguments)**
* Allows a function to accept any number of positional arguments.
* Stored as a tuple.
* Use when you don’t know how many positional inputs will come.

**Syntax**
```
def function_name(*args):
    # args is a tuple
    pass
```
****kwargs (Keyword variable arguments)**
* Allows a function to accept any number of keyword arguments.
* Stored as a dictionary.
* Use when you want flexible named parameters.

**Syntax**
```
def function_name(**kwargs):
    # kwargs is a dictionary
    pass
```
**Combined syntax**
```
def function_name(*args, **kwargs):
    pass
```

**Full parameter order (important)**
```
def func(normal_args, *args, default_args=10, **kwargs):
    pass
```
* Python reads arguments left → right, so:
1. First fixed values
2. Then flexible positional (*args)
3. Then optional defaults
4. Finally flexible keywords (**kwargs)
* "Normal → *args → defaults → **kwargs"

#Constructor
* A constructor is a special method that is automatically called when an object is created. It is used to initialize object attributes (variables).

```
__init__()
```
**Key Features**
* Defined using def __init__(self, ...)
* self refers to the current object
* Automatically executed during object creation
* Used to assign values to object properties

**Types of Constructors**
1. Default Constructor
* Takes no extra parameters (only self)
```
class Test:
    def __init__(self):
        print("Default constructor")
```        
2. Parameterized Constructor
* Takes arguments to initialize values
```
class Test:
    def __init__(self, x):
        self.x = x
```        
**Constructor with Default Values**
```
class Test:
    def __init__(self, x=0):
        self.x = x
```        
**Multiple Constructors in Python?**

* Python does NOT support multiple constructors directly
* Use default arguments or class methods instead
```
class Test:
    def __init__(self, x=None):
        if x is None:
            self.x = 0
        else:
            self.x = x
```
**Constructor Overloading (Workaround)**
* Handled using default values or *args
```
class Test:
    def __init__(self, *args):
        print(args)
```
**Important Points**
* Constructor name is always __init__
* It is not mandatory (Python provides default one)
* Can’t return a value (only None)
* Runs once per object creation


**Destructor (Related Concept)**
* Special method: __del__()
* Called when object is destroyed
```
def __del__(self):
    print("Object destroyed")
```

````python
#Encapsulation

class update:
  def Hide(self,pf,s):
    self.__pf = pf
    self.__s = s
  def show(self):
    print("the pf no is",self.__pf)
    print("the salary is",self.__s)

obj = update()
obj.Hide(1234,50000)
obj.show()
````

````python
class Bank_Details:
  def __init__(self,n,ac,bal):
    self.__n = n
    self.__ac = ac
    self.__bal = bal
    print("Initial balance is",bal)
  def Deposit(self,add):
    self.__bal += add
    print("The balance after deposit is",self.__bal)
  def Withdraw(self,sub):
    print("Withdrawal amount is",sub)
    if self.__bal<sub:
      print("Withdrawal is not possible as insufficient balance",self.__bal)
      print("Withdrawal request terminated...,try again with amount within reach")
    else:
      print("Initiating Withdrawal...")
      self.__bal -= sub
      print("After withdrawal balance is",self.__bal)
      print("Withdrawal complete")

obj = Bank_Details("Ag",1324,0)
obj.Deposit(1000)
obj.Withdraw(2000)
````

````python
#Create two classes: Square → method area() , Rectangle → method area() Call them using Object
class square:
  def area(self,s):
    self.s = s
    print("area of square is",s*s)
class Rectangle(square):
  def area(self,l,w):
    print("area of rectangle is",l*w)

obj = square()
obj.area(2)
obj1 = Rectangle()
obj1.area(2,2)
````

````python
#Create an abstract class Payment with abstract methods pay() and payment_type(). Implement it in CreditCard and UPI class.
from abc import ABC,abstractmethod
class Payment(ABC):
  @abstractmethod
  def pay(self):
    pass
  @abstractmethod
  def payment_type(self):
    pass
class CreditCard(Payment):
  def pay(self):
    print("Paid")
  def payment_type(self):
    print("CC")
class UPI(Payment):
  def pay(self):
    print("Unpaid")
  def payment_type(self):
    print("UPI")

obj1 = UPI()
obj1.pay()
obj1.payment_type()
obj2 = CreditCard()
obj2.pay()
obj2.payment_type()
````

````python
#Create an abstract class Food with abstract methods prepare() and price(). Implement it in Pizza and Burger class.
from abc import ABC,abstractmethod
class Food(ABC):
  @abstractmethod
  def prepare(self):
    pass
  @abstractmethod
  def price(self):
    pass
class Pizza(Food):
  def prepare(self):
    print("peri-peri pizza")
  def price(self):
    print("249")
class Burger(Food):
  def prepare(self):
    print("veg paty burger")
  def price(self):
    print("99")

obj1 = Burger()
obj1.prepare()
obj1.price()
obj2 = Pizza()
obj2.prepare()
obj2.price()
````

````python
#Create an abstract class Transport with abstract methods travel() and fare(). Implement it in Bus and Train class.
from abc import ABC,abstractmethod
class Transport(ABC):
  @abstractmethod
  def travel(self):
    pass
  @abstractmethod
  def fare(self):
    pass
class Bus(Transport):
  def travel(self):
    print("mumbai-pune")
  def fare(self):
    print("225")
class Train(Transport):
  def travel(self):
    print("Mumbai-pune-mumbai")
  def fare(self):
    print("299")

obj1 = Train()
obj1.travel()
obj1.fare()
obj2 = Bus()
obj2.travel()
obj2.fare()
````

````python
#Create an abstract class Employee with abstract methods work() and salary(). Implement it in Manager class.
from abc import ABC,abstractmethod
class Employee(ABC):
  @abstractmethod
  def Work(self):
    pass
  @abstractmethod
  def Salary(self):
    pass
class Manager(Employee):
  def Work(self):
    print("Branch Manager")
  def Salary(self):
    print("60000")

obj1 = Manager()
obj1.Work()
obj1.Salary()
````
